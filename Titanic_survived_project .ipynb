# Removing Outliers Using Zscore

```python
# importing zscore library
from scipy.stats import zscore
```

```python
outliers_features= np.abs(zscore(tit))
threshold=3
print('Outliers:- \n',np.where(outliers_features>3))
```

```python
# z score for index position 13,5 and 40,1
z_score = outliers_features
print('zscore for index position 13 and 5:', z_score.iloc[13,5])
print('zscore for index position 16 and 4:', z_score.iloc[16,4])
print('Samples of zscore index position are >3')

# Removing outliers and storing data in new variable
new_tit =tit[(z_score <3).all(axis =1)]
print('\nBefore removing outliers:', tit.shape)
print('After removing outliers:', new_tit.shape)
```

# Removing Outliers Using IQR

```python
 # find the IQR (Inter quantile rang) to identify outliers 

#1st quantile
q1 = tit.quantile(0.25)

#3rd quantile 
q3 = tit.quantile(0.75)

#IQR
iqr = q3 - q1
```

```python
q3
```

```python
# For Age 
age_high = (q3.Age + (1.5 * iqr.Age))
age_high
```

```python
# checking the index which have higher values (54.5)
np_index_age = np.where(tit['Age']> age_high)
print('Outliers indexes for Age:', np_index_age)
```

```python
# Drop index which we found in the above cell 
tit = tit.drop(tit.index[np_index_age])
tit.shape
tit.reset_index()
```

```python
# For Age 
sibsp_high = (q3.SibSp + (1.5 * iqr.SibSp))
print('\nQ3 value for SipSp:', sibsp_high )

# checking the index which have higher values
np_index_sibsp = np.where(tit['SibSp']> sibsp_high)
print('\nOutliers indexes for SibSp:', np_index_sibsp)

# Drop index which we found in the above cell
tit_sibsp = tit.drop(tit.index[np_index_sibsp])
print('Rows before removing outliers in sibsp:', tit.shape)
print('\nRows after removing outliers in sibsp:', tit_sibsp.shape)
tit=tit_sibsp
tit.reset_index()
```

```python
# For Parch
Parch_high = (q3.Parch + (1.5 * iqr.Parch))
print('Q3 value for Parch:', Parch_high )

# checking the index which have higher values
np_index_parch = np.where(tit['Parch']> Parch_high)
print('\nOutliers indexes for Parch:', np_index_parch)

# Drop index which we found in the above cell
tit_parch = tit.drop(tit.index[np_index_parch])
print('\nRows before removing outliers in Parch:', tit.shape)
print('Rows after removing outliers in Prach:', tit_parch.shape)

tit=tit_parch
tit.reset_index()
```

```python
# For Fare
fare_high = (q3.Fare + (1.5 * iqr.Fare))
print('Q3 value for Fare:', fare_high )

# checking the index which have higher values
np_index_fare = np.where(tit['Fare']> fare_high)
print('\nOutliers indexes for Fare:', np_index_fare)

# Drop index which we found in the above cell
tit_fare = tit.drop(tit.index[np_index_fare])
print('\nRows before removing outliers in Fare:', tit.shape)
print('Rows after removing outliers in Fare:', tit_fare.shape)

tit=tit_fare
tit.reset_index()
```

```python
# For low Age 
age_low = (q1.Age - (1.5 * iqr.Age))
print('Q1 value for Age:', age_low )

# checking the index which have higher values
np_index_age = np.where(tit['Age']< age_low)
print('\nOutliers indexes for Age:', np_index_age)

# Drop index which we found in the above cell
tit_age = tit.drop(tit.index[np_index_age])
print('\nRows before removing outliers in Age:', tit.shape)
print('Rows after removing outliers in Fare:', tit_age.shape)

tit=tit_age
tit.reset_index()
```

# Scaling the Data

```python
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
print('After scaling data:\n', x_scaled)
print('\nColumn of the data:', x_scaled.shape[1])
```

 Lets find the features that are correlated with other features using Variance inflation Factor(vif)

```python
vif = pd.DataFrame()
vif['VIF']= [variance_inflation_factor(x_scaled, i)for
            i in range (x_scaled.shape[1])]
vif ['Features'] = x.columns 
print(vif)
```

All the vif values are less than < 5, that means we can use all features to build our model since there is no mutilcollinearity issues amongs features 

# Spliting our data set into train and test

```python
from sklearn.model_selection import train_test_split, GridSearchCV
x_train, x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.25,random_state=355)
print('Preprocessing steps Completed ')
```

# Model Building 

```python
# Importing the neccessary Classification Models 
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,roc_auc_score
from sklearn.model_selection import cross_val_score
```

# Logistic Regression Algorithm 

```python
#initiate LogesticRegression and Train Model
LR = LogisticRegression()
LR.fit(x_train,y_train)
            
```

```python
def metric_score(LR, x_train, x_test, y_train, y_test, train=True):
    if train:
        
        y_pred = LR.predict(x_train)
        
        print("\n============Train Result=====================")
        
        print(f"Accuracy Score: {accuracy_score(y_train, y_pred)*100:.2f}%")
        
    elif train==False:
        pred = LR.predict(x_test)
        
        print("\n==================Test Result==================")
        print(f"Accuracy Score:{accuracy_score(y_test, pred) * 100:.2f}%")
        print('\n')
        
        print (f"The CV score is: {cross_val_score(LR, x, y, cv=5).mean()* 100:.2f}%")
        print ("Difference between Accuracy_Score and CV score is: ",((accuracy_score(y_test, pred)*100)-((cross_val_score(LR, x, y, cv=5).mean())*100)))
        print ('\n')
        
        
        print("Confusion Matrix is: \n", confusion_matrix(y_test, pred))
        print ('\n')
        print('\n\n Test Classification Report \n',classification_report(y_test, pred, digits=2)) #confidence/accuracy
        
```

```python
# callING the function and pass dataset to check 

metric_score(LR,x_train,x_test,y_train,y_test,train=True)# For Training Score 
metric_score(LR,x_train,x_test,y_train,y_test,train=False)# for Test Score 


```
